```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as matimg
from keras.models import load_model
import import_ipynb
from vae import latent2im,im2latent
```

```python
img = Image.fromarray(matimg.imread('2.jpg'))
img = np.array(img.resize([64, 64]))

plt.imshow(img)
plt.show()

img = img.astype('float32') / 255.
img = img.reshape((1, np.prod(img.shape[0:])))
```

```python
encoder = load_model('encoder.h5')
decoder = load_model('decoder.h5')
comp = im2latent(encoder,img)
print(comp)
```

```python
im2 = latent2im(decoder,comp) 
plt.imshow(im2)
plt.show()
```

```python
#to generate a new random image
encoded_rep = np.random.rand(1,20)
print(encoded_rep)
plt.figure(figsize=(10, 2))
im = latent2im(decoder,encoded_rep) 
plt.imshow(im)
plt.show()
```
